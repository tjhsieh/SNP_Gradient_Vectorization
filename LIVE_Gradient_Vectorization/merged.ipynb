```python
import os
import xml.etree.ElementTree as ET
import shutil
```

```python
# collect and copy path svg to file_direct
path_num = 1 
file_direct = os.path.join("resource/flower/path"+str(path_num),"")
log_direct = "log/flower/path10/sunflowers_rendered_single_stroke_"
if not os.path.isdir(file_direct):
    os.mkdir(file_direct)
for stroke_num in range(0,495) :
    stroke_name = ""
    num = str(stroke_num+1)
    num_of_zeros = 4 - len(num)
    for i in range(num_of_zeros) :
        stroke_name += '0'
    stroke_name += num
    src_filname = log_direct +  stroke_name + "/output-svg/" + str(path_num) + ".svg"
    dst_filname = file_direct + stroke_name + ".svg"
    shutil.copyfile(src_filname, dst_filname )
```

```python
ET.register_namespace('', 'http://www.w3.org/2000/svg')

allFileList = os.listdir(file_direct)

svg = ET.Element('svg', width='512', height='512')
i = 1

for file in allFileList:
    print(file)
    group = ET.Element('g')
    num = str(i)
    if len(num) == 1 :
        num = '000' + num
    if len(num) == 2:
        num = '00' + num
    if len(num) == 3:
        num = '0' + num
    group.set('id',"stroke_" + str(num))
    tree1 = ET.parse(os.path.join(file_direct,file))
    root1 = tree1.getroot()
    for path in root1.iter('{http://www.w3.org/2000/svg}path'):
        fill_attrib = path.attrib.get('fill')
        rgb_values = fill_attrib.split(',')
        r = int(rgb_values[0][4:])
        g = int(rgb_values[1])
        b = int(rgb_values[2][:-1])
        brightness = (r + g + b) / 3
        if brightness < 240:
            group.append(path)
    svg.append(group)
    i+=1
tree = ET.ElementTree(svg)   
tree.write( file_direct + "merged_" + str(path_num) + ".svg" )
```
