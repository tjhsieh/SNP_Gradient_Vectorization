```python
import numpy as np
import PIL
import PIL.Image
import cv2
import matplotlib.pyplot as plt
from skimage.metrics import structural_similarity as ssim
from skimage.metrics import mean_squared_error
```

```python
# Convert the images to grayscale if needed
gt = cv2.imread("C:/Users/aaaws/Desktop/LIVE/web/result_png/final.png",0)
gt_norm = gt.astype(np.float64) /255.0

inkscape = cv2.imread("C:/Users/aaaws/Desktop/LIVE/web/result_png/Inkscape.png",0)
inkscape_norm = inkscape.astype(np.float64) /255.0

live = cv2.imread("C:/Users/aaaws/Desktop/LIVE/web/result_png/final_600.png",0)
live_norm = live.astype(np.float64) /255.0

merged_1 = cv2.imread("C:/Users/aaaws/Desktop/LIVE/web/result_png/merged_1.png",0)
merged_1_norm = merged_1.astype(np.float64) /255.0

merged_4 = cv2.imread("C:/Users/aaaws/Desktop/LIVE/web/result_png/merged_4.png",0)
merged_4_norm = merged_4.astype(np.float64) /255.0

gradient = cv2.imread("C:/Users/aaaws/Desktop/LIVE/web/result_png/merged_hsv_gradient.png",0)
gradient_norm = gradient.astype(np.float64) /255.0

# Compute the SSIM between the two images
ink_score = ssim(gt_norm, inkscape_norm, win_size=3,K2=0.03, data_range=inkscape_norm.max() - inkscape_norm.min())
ink_mse = mean_squared_error(gt_norm,inkscape_norm)
print("Inkscape")
print("SSIM score:", "{:.5f}".format(ink_score) )
print("MSE score:", "{:.5f}".format(ink_mse))

live_score = ssim(gt_norm, live_norm, win_size=3, K2=0.03, data_range=live_norm.max() - live_norm.min())
live_mse = mean_squared_error(gt_norm,live_norm)
print("Live")
print("SSIM score:", "{:.5f}".format(live_score) )
print("MSE score:", "{:.5f}".format(live_mse))

# Compute the SSIM between the two images
merged_1_score = ssim(gt_norm, merged_1_norm, win_size=3,K2=0.03,data_range=merged_1_norm.max() - merged_1_norm.min())
merged_1_mse = mean_squared_error(gt_norm,merged_1_norm)
print("merged_1")
print("SSIM score:", "{:.5f}".format(merged_1_score) )
print("MSE score:", "{:.5f}".format(merged_1_mse))

# Compute the SSIM between the two images
merged_4_score = ssim(gt_norm, merged_4_norm, win_size=3,K2=0.03,data_range=merged_4_norm.max() - merged_4_norm.min())
merged_4_mse = mean_squared_error(gt_norm,merged_4_norm)
print("merged_4")
print("SSIM score:", "{:.5f}".format(merged_4_score) )
print("MSE score:", "{:.5f}".format(merged_4_mse))

# Compute the SSIM between the two images
gradient_score = ssim(gt_norm, gradient_norm, win_size=3,K2=0.03, data_range=gradient_norm.max() - gradient_norm.min())
gradient_mse = mean_squared_error(gt_norm,gradient_norm)
print("merged_hsv")
print("SSIM score:", "{:.5f}".format(gradient_score) )
print("MSE score:", "{:.5f}".format(gradient_mse))
```
